```python
scientist = culled_df2[culled_df2[df2.columns[0]].str.contains('scientist|Psych', regex = True)]
scientist
```

```python
sci_emp2020 = scientist[df2.columns[3]].sum()
print(sci_emp2020)
sci_emp2030 = scientist[df2.columns[4]].sum()
print(sci_emp2030)
sci_change = sci_emp2030 - sci_emp2020
print(sci_change)
sci_inc = sci_change / sci_emp2020 * 100
print(sci_inc)
```

```python
comp_sci = culled_df2[culled_df2[df2.columns[0]].str.contains('Computer occupations')]
comp_sci_sub = culled_df2[culled_df2[df2.columns[0]].str.contains('Computer support')]

comp_sci2020 = comp_sci[df2.columns[3]].sum()
comp_sci2030 = comp_sci[df2.columns[4]].sum()

comp_scisub2020 = comp_sci_sub[df2.columns[3]].sum()
comp_scisub2030 = comp_sci_sub[df2.columns[4]].sum()

comp_sci2020 = comp_sci2020 - comp_scisub2020
print(comp_sci2020)
comp_sci2030 = comp_sci2030 - comp_scisub2030
print(comp_sci2030)

comp_sci_change = comp_sci2030 - comp_sci2020
print(comp_sci_change)

compsci_inc = comp_sci_change / comp_sci2020 * 100
print(compsci_inc)
```
